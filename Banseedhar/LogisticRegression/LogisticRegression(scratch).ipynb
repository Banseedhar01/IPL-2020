```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as pt
```

```python
data=pd.read_csv("Social_Network_Ads.csv")
data.head()
```

```python
data.shape

```

```python
ismale=pd.get_dummies(data['Gender'],drop_first=True)
data=pd.concat([data,ismale],axis=1)
data.head()
```

```python
data.corr()
```

```python
x=(data[['Age','EstimatedSalary',]]).values
y=(data['Purchased']).values
```

```python
#normalization

x=(x-np.mean(x))/np.std(x)

#biasing
x = np.column_stack((np.ones((x.shape[0],1)),x))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=20)
```

```python
def sig(z):
    return 1/(1+np.exp(-z))
```

```python
def loss(yp,y):
    return (-1/y.shape[0])*(y.T@np.log(yp)+(1-y).T@np.log(1-yp))
```

```python

B = np.zeros((x_train.shape[1],1))
print(B)
alpha =3.4
L =[]
y_train=y_train.reshape(x_train.shape[0],-1)
for i in range(2000):
  k = x_train@B
  z = sig(k)
  dB = (1/x_train.shape[0])*np.dot(x_train.T,(z-y_train))
  B = B-dB*alpha
  L.append(loss(z,y_train))
print(L)
```

```python
print(B)
```

```python
k=x_test@B
y_pre=sig(k)
for i in range(y_pre.shape[0]):
  if y_pre[i]>=0.53:
    y_pre[i]=1
  else :
    y_pre[i]=0
y_pre=y_pre.reshape(-1)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pre)
print(cm)
```
