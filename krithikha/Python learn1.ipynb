```python
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv('kc_house_data.csv')
df.head()
```

```python
#finding which feature has most impact on price
imp=df.corr()
seaborn.heatmap(imp,cmap="YlGnBu")
```

```python
#We can see sqft_living is dark in price coloum indicating high impact.So we consider that a good feature
#feature_cols = sqft_living 
```

```python
x = df['sqft_living'].values
y=df['price'].values

#splitting the data
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression

#creating training set and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.221,random_state=2)

x_train= x_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
print('Intercept: {}'.format(lr.intercept_))
print('Coefficient: {}'.format(lr.coef_))
```

```python
y_pred= -40103.63411459 + 279.03569292 * x
x_test=x_test.reshape(-1,1)
prediction=lr.predict(x_test)
```

```python
# mean squared error MSE = 1/n(y(tested)-y(prediction))^2
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,lr.predict(x_test))
np.sqrt(mse)
```

```python
#Final score
lr.score(x_test,y_test)
```

```python
plt.scatter(x,y,color="green")
plt.xlabel('sqft_living')
plt.ylabel('Price')
plt.plot(x, y_pred, color='black')
plt.show()
```
